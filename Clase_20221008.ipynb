# **CICLOS**

* Ciclo *for*: Nosotros a priori la cantidad de iteraciones que vamos a hacer.
* Ciclo *while*: Nosotros no conocemos la cantidad de veces que vamos a realizar una acción. La acción se va a ejecutar hasta que se cumpla o no una condición

```python
for num in range(1, 11, 1): # range(inicial, final, pasos o incremento)
  print(num)
```

```python
frutas = ['Banano','Mango','Papaya','Kiwi']
print(frutas[0])
print(frutas[1])
print(frutas[2])
print(frutas[3])
```

```python
for fruta in frutas:
  print(fruta)
```

```python
for fruta in frutas:
  print(fruta)
  if fruta == 'Mango':
    break
```

```python
for fruta in frutas[::-1]:
  print(fruta)
  if fruta == 'Mango':
    break
```

```python
for color in ['Rojo','Amarillo','Verde','Fucsia']:
  print(color)
```

```python
for fila in [1, 2, 3]:
  print("",)
  for columna in [1, 2]:
    print("*", end = '')
```

```python
for fila in range(0, 15, 1):
  print("")
  for columna in range(0, fila + 1, 1):
    print('*', end = '')
```

```python
for fila in range(0, 10, 1):
  for columna in range(0, 15, 1):
    print("*", end = '')
  print("")
```

```python
for fila in range(0, 2, 1):
  print("")
  for columna in range(0, 15, 1):
    print('*', end = '')
```
