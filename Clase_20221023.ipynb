### Manejo de errores en Python
* Para manejar en Python, usamos un bloque de código que se llama try/catch, los cuales, debemos implementar en la parte de nuestro código que nosotros sospechemos pueda fallar.

```python
# Este try catch no hace sentido
try:
  result = 2 + 2
except Exception as e:
  print(e)

print(result)
```

```python
num2 = int(input("Por favor ingrese un número: "))
try:
  result = 2 / num2
  print(f"El resultado de 2 / {num2} es = {result}")
except Exception as e:
  print(f"Ha ocurrido un error, no se permite las divisiones por 0.")
  print(e)
```

```python
def generar_grupo(edad_flor):
  if edad_flor >= 15 and edad_flor <= 30:
    return 'Edad entre 15 y 30'
  elif edad_flor > 30 and edad_flor <= 75:
    return 'Edad entre 31 y 75'
  else:
    return 'Edad mayor a 75'
```
