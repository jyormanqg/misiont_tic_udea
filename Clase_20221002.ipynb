**Parametro** Es una variable o un valor que se le va a enviar a la función para que esta pueda realizar sus tareas.

**Ejercicio**

Construya una función que se encargue de calcular el valor de un producto, en base, a si el producto tiene o no IVA.
La función va a recibir como parametros: El valor del producto y si tiene o no IVA (puede ser una variable que tenga 'si' o 'no'. (tomen el IVA como 0.19)

La función debe imprimir lo siguiente:
* El producto tiene un costo de 335.55, y con IVA es de 399.3045
* El producto tiene un costo de 500, y no tiene IVA.

```python
def ejercicio(valor_producto, tiene_iva):
  iva = 0.19
  if tiene_iva == 'si':
    valor_total = valor_producto + (valor_producto * iva)
    print(f"El producto tiene un costo de {valor_producto}, y con IVA es de {valor_total}")
  else:
    print(f"El producto tiene un costo de {valor_producto}, y no tiene IVA")
```

```python
ejercicio(3000, 'si')
```
