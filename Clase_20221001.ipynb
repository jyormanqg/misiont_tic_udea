## Funciones

Una función es una porción de código que se encarga de realizar una o varias tareas especificas, que puede recibir o no parametros, y puede retornar o no datos.

La sintaxis para crear una función es la siguiente:

```
def nombre_funcion(parametros (opcional)): # Definición de la función
    print("Hola, esta es una función") # Código interno
    return 0 # Return
```

Las funciones constan de varias partes:
1. Definición de la función, que se hace con la palabra **def nombre_funcion():**
2. El código interno de la función.
3. El return (opcional)
4. El llamado a la función.

````python
def mi_primera_funcion(): # Definición de la función
  print("Hola, esta es mi primer función en Python") # Código interno de la función
  print("Este es otro mensaje")
````

````python
# LLamado a la función
mi_primera_funcion() # Siempre van los parentesis vacios o no, para llamar la función
mi_primera_funcion() # Siempre van los parentesis vacios o no, para llamar la función
mi_primera_funcion() # Siempre van los parentesis vacios o no, para llamar la función
````

**Los parametros** de las funciones, son valores que reciben las funciones, y puede trabajar con ellas, o simplemente mostrarlas en pantalla o no hacer nada.

Los parametros van dentro de los parentesis **(parametro1, parametro2, parametro3, ...)**

````python
def sumar(num1, num2):
  resultado = num1 + num2
  return resultado
````

````python
res = sumar(10, 15)
print(f"El resultado fue: {res}")
````
