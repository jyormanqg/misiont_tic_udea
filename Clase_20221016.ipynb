```python
personas = {
    1 : {
        'nombre'    : 'Yorman',
        'edad'      : 28,
        'nro_hijos' : 0
    },
    2 : {
        'nombre'    : 'Pepito',
        'edad'      : 15,
        'nro_hijos' : 0
    },
    3 : {
        'nombre'    : 'Carolina',
        'edad'      : 33,
        'nro_hijos' : 5
    },
    4 : {
        'nombre'    : 'Carlitos',
        'edad'      : 11,
        'nro_hijos' : 0
    },
    5 : {
        'nombre'    : 'Fulanita',
        'edad'      : 35,
        'nro_hijos' : 2
    },
}
```

```python
# 1. Identificar la cantidad de personas en la lista
# 2. Cantidad de personas con edad mayor o igual que 15
# 3. Cantidad de personas con hijos
# 4. Suma total de hijos de las personas
# 5. Número unico de cantidad de hijos 
```

```python
nro_personas           = len(personas)
nro_peronas_mayor15    = 0
cnt_personas_con_hijos = 0
suma_total_hijos       = 0
lista_nro_hijos        = [] # Inicializamos la lista vacia para irla llenando con los numeros de hijos de cada persona
nro_unico_cnt_hijos    = 0

for key, persona in personas.items():
  if persona['edad'] >= 15:
    nro_peronas_mayor15 += 1

  if persona['nro_hijos'] > 0:
    cnt_personas_con_hijos += 1
    suma_total_hijos += persona['nro_hijos']
  
  lista_nro_hijos.append(persona['nro_hijos'])


lista_nro_hijos = set(lista_nro_hijos)
nro_unico_cnt_hijos = len(lista_nro_hijos)


print(f"1. número de personas: {nro_personas}")
print(f"2. Cantidad de personas con edad mayor o igual que 15: {nro_peronas_mayor15}")
print(f"3. Cantidad de personas con hijos: {cnt_personas_con_hijos}")
print(f"4. Suma total de hijos de las personas: {suma_total_hijos}")
print(f"5. Número unico de cantidad de hijos: {nro_unico_cnt_hijos}")
```
