## Aclaración de como usar el Axis = 0 y el Axis = 1

* Axes are defined for arrays with more than one dimension. A 2-dimensional array has two corresponding axes: the first running vertically downwards across rows (axis 0), and the second running horizontally across columns (axis 1).

```python
frutas          = ['Banano','Mango','Banano','Papaya','Mango','Corozo','Mango','Mango','Pera']
frutas_deseadas = ['Mango','Corozo']
frutas
```

```python
cont_bananos  = 0
cont_mango    = 0
cont_deseadas = 0

for fruta in frutas:
  if fruta == 'Banano':
    cont_bananos += 1
  elif fruta == 'Mango':
    cont_mango += 1

  if fruta in frutas_deseadas:
    cont_deseadas += 1

print(cont_bananos)
print(cont_mango)
print(cont_deseadas)
```

```python
cnt_grupos_ventas = len(lista_ventas) / 2
```

```python
array_ventas = np.array_split(lista_ventas, cnt_grupos_ventas)
array_ventas
```

```python
mantequilla = 0
leche       = 0
huevos      = 0

for prod in array_ventas:
  if prod[0] == 'mantequilla':
    mantequilla += int(prod[1])
  elif prod[0] == 'leche':
    leche += int(prod[1])
  elif prod[0] == 'huevos':
    huevos += int(prod[1])
```
